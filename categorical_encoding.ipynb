## One Hot Encoding with multiple categories

```python
for col in data.columns:
    print(col, ":", len(data[col].unique()), 'labels')
```

```python
pd.get_dummies(data,drop_first=True).shape
```

```python
data.X2.value_counts().sort_values(ascending=False).head(20)
```

```python
top_10 = [x for x in data.X2.value_counts().sort_values(ascending=False).head(10).index]
top_10
```

```python
for label in top_10:
    data[label] = np.where(data['X2']==label,1,0)
data[['X2']+top_10].head(40)
```

```python
def on_hot_encoding(df):
    
    for col in df.columns:
        if len(data[col].unique()) >= 10:
            print("1")
            top_10 = [x for x in data[col].value_counts().sort_values(ascending=False).head(10).index]
            print(top_10)
            for label in top_10:
                df[col + '_' + label] = df[col].apply(lambda x: 1 if x == label else 0)
            print(df)
        else:
            print("2")
            encoded_values = pd.get_dummies(df[col], prefix=col)
            print(encoded_values)
            encoded_values = encoded_values.map(lambda x: 1 if x == True else 0)
            print(encoded_values)
            df = pd.concat([df,encoded_values],axis=1)
            df = df.drop(columns=[col])
            print(df)
    return df
data = pd.read_csv('data/mercedes-benz.csv', usecols=["X1","X2","X3","X4","X5","X6"])
data = on_hot_encoding(data)
```

```python
data = data.drop(columns=["X1","X2","X5","X6"])
data
```

## Count / Frequency Encoding

```python

data = pd.read_csv('data/mercedes-benz.csv', usecols=["X1","X2","X3","X4","X5","X6"])
for col in data.columns:
    print(col, ":", len(data[col].unique()), 'labels')
```

```python
data.X2.value_counts().to_dict()
```

```python
def count_frequency_encoding(df):
    for col in df.columns:
        if df[col].dtype == 'object':
            df_frequency_map = df[col].value_counts().to_dict()
            df[col] = df[col].map(df_frequency_map)
    return df

data = count_frequency_encoding(data)
data
```

## Ordinal / Label Encoding

```python
import pandas as pd
import datetime
```

```python
df_base = datetime.datetime.today()
df_date_list = [df_base - datetime.timedelta(days=x) for x in range(0, 20)]
df = pd.DataFrame(df_date_list)
df.columns = ['day']
df
```

```python
df['day_of_week'] = df['day'].dt.day_name()
df.head()
```

```python
df['day_ordinal'] = df['day'].dt.weekday
df.head()
```
